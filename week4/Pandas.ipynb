## Pandas

```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('BreadBasket_DMS.csv')
df
```

```python
df['Item'].value_counts().plot(kind='bar',figsize=(20,4))
```

```python
from pytz import datetime as dt
```

```python
datetime = df.Date+" "+df.Time
datetime
```

```python
df['datetime']=pd.to_datetime(datetime,format='%Y-%m-%d %H:%M:%S')
```

```python
df.datetime[0]
```

```python
dt.datetime.now()
```

```python
df['hours'] = df.datetime.apply(lambda x : x.hour)
```

```python
df['hours'].value_counts(sort=False).plot(kind='bar')
```

```python
df = df.sort_values(by=['hours'])
```

```python
df['hours']
```

```python
df['hours'].value_counts().sort_index().plot(kind='line',figsize=(15,10)).set_xticks(range(24))
```

```python
new_df = df[df['Item']=='Coffee']
```

```python
df.loc[df['Item']== 'Coffee'].hist()
```
